Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a,b,c = map(float, input().split())
p=(a+b+c)/2
s=(p*(p-a)*(p-b)*(p-c))**0.5
if a+b>c and a+c>b and b+c>a:
    if a==b!=c or b==c!=a or c==a!=b:
        print(f"Tam giác cân có diện tích,chu vi lần lượt là: {round(s)},{round(p)}")
    elif a==b==c:
        print(f"Tam giác đều có diện tích,chu vi lần lượt là: {round(s)},{round(p)}")
    elif a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
        print(f"Tam giác vuông có diện tích,chu vi lần lượt là: {round(s)},{round(p)}")
    else:
        print(f"Tam giác thường có diện tích,chu vi lần lượt là: {round(s)},{round(p)}")
else:
    print(f"Khong la ba canh cua mot tam giac")


    
```

Bài 8: Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n


```python
a,b,m,c,d,n = map(float,input().split())
D = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c
if D != 0:
    x = Dx/D
    y = Dy/D
    print(f"Hệ phương trình có hệ duy nhất:x={x:.2f},y={y:.2f}")
elif D == 0 and Dx == 0 and Dy == 0: 
    print("Hệ phương trình có vô số nghiệm")
else:
    print("Hệ phương trình vô nghiệm")

```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x,y,r = map(float, input().split())
a,b = map(float, input().split())
kc = ((a-x)**2+(b-y)**2)**0.5
if kc == r:
    print("A thuộc đường tròn")
else:
    print("A không thuộc đường tròn")

```
