Thuật toán Shoelace (Dây giày)

```python
def tinh_dien_tich(quad):
    n = len(quad) # Thường là 4 cho tứ giác
    tong_xuoi = 0
    tong_nguoc = 0
    
    for i in range(n):
        # Lấy đỉnh hiện tại (i) và đỉnh kế tiếp (j)
        # j = (i + 1) % n để khi đến đỉnh cuối nó tự quay về đỉnh đầu
        j = (i + 1) % n
        
        tong_xuoi += quad[i][0] * quad[j][1]
        tong_nguoc += quad[i][1] * quad[j][0]
        
    dien_tich = 0.5 * abs(tong_xuoi - tong_nguoc)
    return dien_tich

# Chạy thử với ví dụ của Cường
quad = [(0,0), (4,0), (4,3), (0,3)]
print(f"Diện tích tứ giác là: {tinh_dien_tich(quad)}") 
# Kết quả sẽ ra 12.0 (4 * 3 = 12, quá chuẩn!)
```

Làm tròn


```python
number = 0
ndigits =0
round(number, ndigits)# Cách phổ biến nhất.
#round(3.567, 2) = 3.57round(3.567) = 4
# Lưu ý "lạ": Python làm tròn đến số chẵn gần nhất khi ở mức 0.5 (ví dụ round(2.5) ra 2, round(3.5) ra 4).

#Thư viện math (Làm tròn tuyệt đối):
math.ceil(3.1) = 4#(Luôn làm tròn lên).
math.floor(3.9) = 3#(Luôn làm tròn xuống).
math.trunc(3.9) = 3#(Cắt bỏ phần thập phân, giữ nguyên phần nguyên).

#Sử dụng f-string (Cách hiện đại nhất)
pi = 3.14159265
print(f"{pi:.2f}")  # Lấy 2 chữ số thập phân -> 3.14
print(f"{pi:.0f}")  # Không lấy chữ số thập phân nào -> 3

#Sử dụng phương thức .format() (Kiểu cũ hơn chút):
print("{:.3f}".format(pi)) # -> 3.142

#Sử dụng toán tử % (Kiểu giống ngôn ngữ C):
print("%.2f" % pi) # -> 3.14

#Làm tròn với định dạng số lớn (Dấu phẩy ngăn cách hàng nghìn)
so_tien = 1234567.891
# Vừa lấy 2 chữ số thập phân, vừa ngăn cách hàng nghìn bằng dấu phẩy
print(f"{so_tien:,.2f}") 
# Kết quả: 1,234,567.89
```
