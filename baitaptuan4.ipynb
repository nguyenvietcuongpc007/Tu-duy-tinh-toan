[a, b]. Mỗi dòng là một bộ test (a b)."


```python
def snt(n):
    if n<2: return False
    for i in range(2,int(n**0.5) + 1):
        if n % i == 0: return False
    return True
while True:
    try:
        line = input().strip()
        if not line:
            break
        tong = 0
        a,b = map(int, line.split())
        for i in range(a,b+1):
            if snt(i) == True:
                tong += i
        print(f'Với khoảng {a} đến {b}thì tổng số nguyên tố là{tong}')

    except EOFError:
        break

```

Bài 8:ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 


```python
def reverse(n):
    res = 0
    while n != 0:
        res = res*10 + n%10
        n //= 10
    return res
    #return int(str(n)[::-1])
n = int(input())
so_buoc = 0
while n != reverse(n):
    n = n + reverse(n)
    so_buoc +=1
print(n,so_buoc)
```

Bài 10:"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 

```python
def co(n):
    dem = 1
    while n > 1:
        if n % 2 == 0:
            n = n//2
            dem += 1
        else:
            n = 3*n +1
            dem += 1
    return dem
n = int(input())
max = 0
so = 0
for i in range(1,n+1):
    collatz = co(i)
    if collatz>max:
        max = i
        so = collatz
print(max,so)
```

```python
# Sổ tay lưu độ dài đã tính (Mặc định số 1 có độ dài là 1)
memo = {1: 1}

def co_fast(n):
    # Nếu n đã có trong sổ tay, trả về luôn không cần tính lại
    if n in memo:
        return memo[n]
    
    # Tính bước tiếp theo dựa trên chẵn/lẻ
    if n % 2 == 0:
        res = 1 + co_fast(n // 2)
    else:
        res = 1 + co_fast(3 * n + 1)
    
    # Ghi vào sổ tay trước khi trả về
    memo[n] = res
    return res

# Đoạn tìm max giữ nguyên logic của Cường
n = int(input())
max_len = 0
best_x = 0

for i in range(1, n + 1):
    curr_len = co_fast(i)
    if curr_len > max_len:
        max_len = curr_len
        best_x = i

print(best_x, max_len)
```

Bài 13:[ClosePairNumber]
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
def uoc(n):
    tong = 1
    for i in range(2,int(n**0.5)+1):
        if n % i == 0:
            tong += i
            j = n//i
            if j != i:
                tong += j
    return tong
a,b = map(int, input().split())
if uoc(a) == b and uoc(b) == a:
    print("true")
else:
    print("false")
```

Bài 14:[GCD]
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n. 


```python
def uocm(n):
    A=[]
    for i in range(1,int(n**0.5)+1):
        if n % i == 0:
            A.append(i)
            j = n//i
            if j != i:
                A.append(j)
    return A
m,n  = map(int,input().split())
max = 0
for i in range(1,int(m**0.5)+ 1):
    if n % i == 0:
        if i in uocm(n):
            max = i
        j = n//i
        if j != i:
            if i in uocm(n):
                max = i
print(max)
```
