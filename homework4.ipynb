Bài 3:Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
def uoc(n):
    A = []
    for i in range(1,n+1):
        if n % i ==0:
            A.append(i)
    return A
a,b = map(int,input().split())
b = uoc(b)
for i in uoc(a):
    if i in b:
        print(i,end = ' ')
```

Bài 7:Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
le = 0 
chan = 0
while n != 0:
    a = n%10
    if a%2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print(le,chan)
```

Bài 8:
Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 

```python
n = int(input())
s = 0
k = 0
while s + (k+1) < n:
    k += 1
    s += k
print(s, k)
```

Bài 9: Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 


```python
a = int(input())
s = 0
n = 0
while s <= a:
    n += 1
    s += 1/n
print(s,n)
```

Bài 10:Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
max_val = None
min_val = None
while True:
    a = int(input())
    if a == - 1:
        break
    elif max_val is None or min_val is None :
        max_val = a
        min_val = a
    else :
        if a > max_val:
            max_val = a
        elif a < min_val:
            min_val =a
print(max_val,min_val)
```

Bài 11:Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt

```python
a = int(input())
d = 1
s = 1   
while s <= a:
    d,s = s ,s+d
print(d)
```

Bài 12:hập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for trên index; KHÔNG dùng hàm split(), len(

```python
dem = 0
n = input()
i = True
for char in n:
    if char != ' ':
        if i:
            dem += 1
            i = False
    else:
        i = True
print(dem)
```

Bài 13:Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi
Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
s = input().strip()

for i in range(len(s)):
    if s[i] == " ":
        break   
    print(s[i], end="") 
```

Bài 14:Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính
tổng 3 số nguyên đó

Bài 15:Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số

```python
hoa = 0
thuong = 0
so = 0
a = input()
for i in a:
    if i.isupper():
        hoa += 1
    elif i.islower():
        thuong += 1
    elif i.isdigit():
        so += 1
print(hoa,thuong,so)
```
