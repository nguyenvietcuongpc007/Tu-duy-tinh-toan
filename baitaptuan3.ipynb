Bài 11:[Phân loại tam giác]
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường"

```python
a,b,c = map(int,input().split())
if a < b + c and b < a + c and c < b + a:
    if a == b == c:
        print("Tam giác đều")
    elif a == b != c or a != b == c or a == c != b:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải là tam giác")
```

Bài 13:[Tính tiền điện]
Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh 


```python
n = int(input())
if n <=50:
    print(n*1500)
if n <=100 and n > 50:
    print(50*1500 + (n-50)*2000)
if n > 100:
    print(50*1500 + 50*2000 + (n-100)*3000)
```

Bài 14:[Giải phương trình bậc nhất]
Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra nghiệm của phương trình, kết quả làm tròn đến số
thập phân thứ 2. In ra "Vô nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng. 


```python
a,b = map(float,input().split())
if a == 0 and b==0:
    print("Vô số nghiệm")
elif a==0 and b!=0:
    print("Vô nghiệm")
else:
    print("{:.2f}".format(-b/a))
```

Bài 16:[Rounding]
Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm
sẵn có của pyton). Các kết quả cách nhau một dấu cách

```python
x = float(input())
tron_xuong = x // 1
if x % 1 == 0:
    tron_len = x // 1
else:
    tron_len = x // 1 + 1
if x % 1 >= 0.5:
    gan_nhat = x // 1 + 1
else:
    gan_nhat = x // 1
print(f"{tron_len:.0f} {tron_xuong:.0f} {gan_nhat:.0f}")


```

Bài 17:Cấp số nhân
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng
thứ tự a, b, c, d hay không?
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs
d.

```python
a,b,c,d = map(float,input().split())
p = b/a
if c/b==p and d/c==p:
    print("Là cấp số nhân")
else:
    print("NO")
```
