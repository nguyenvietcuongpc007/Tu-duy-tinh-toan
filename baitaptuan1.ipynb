Bài 4Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng
6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
• Các điểm a1, b1, c1 được nhân hệ số 1,
• Các điểm a2, b2 được nhân hệ số 2,
• Điểm a3 được nhân hệ số 3.
Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10
Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6
đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được
phân tách nhau bởi duy nhất 1 dấu cách.
Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân.


```python
a1, b1, c1, a2, b2, a3 = map(float, input().split())
average = ((a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3) / 10
print(f"{average:.1f}")
```

Bài 8Fahrenheit, hay độ F, là một thang nhiệt độ được đặt theo tên nhà vật lý người Đức Daniel
Gabriel Fahrenheit (1686–1736). Dựa vào nhiệt độ Celcius được nhập từ bàn phím, hãy
chuyển sang độ F theo công thức: F = 9/5*C + 32 và in ra màn hình kết quả. Kết quả làm
tròn đến hai chữ số thập phân đằng sau dấu phẩy.

```python
celsius = float(input("Enter Celsius: "))
fahrenheit = (9/5) * celsius + 32
print(f"{fahrenheit:.2f}")
```

Bài 9Sơn Tùng mua một chiếc đồng hồ Movado trên trang mua sắm trực tuyến Amazon. Chiếc
đồng hồ có giá x USD (x nhập từ bàn phím). Bởi Amazon không vận chuyển trực tiếp về
Việt Nam, Đạt phải trả 10 USD cho phí vận chuyển từ một nước trung gian. Khi vận chuyển
về Việt Nam, Đạt được thông báo sẽ phải trả thêm 30% thuế xuất nhập cảnh và 10% thuế
VAT cho chiếc đồng hồ nói trên. Viết chương trình tính tổng số tiền mà Đạt phải trả để mua
được chiếc đồng hồ trên (USD). Kết quả làm tròn đến hai chữ số thập phân.


```python
x = float(input("Enter watch price (USD): "))
shipping = 10
import_tax = 0.30 * x
vat = 0.10 * x
total = x + shipping + import_tax + vat
print(f"Total to pay: {total:.2f} USD")
```
