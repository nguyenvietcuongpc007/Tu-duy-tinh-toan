Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và chiều cao
(kiểu số thực) với hai phương thức:tính diện tích bề mặt (diện
tích toàn phần) và thể tích của hình trụ
Đầu vào
Đâu vào từ bàn phím gồm một dòng duy nhất chứa bán kính và
chiều cao của hình trụ, cách nhau bởi một dấu cách.
Đầu ra
In ra màn hình diện tích bề mặt và thể tích của hình trụ (sử
dụng lại những hàm vừa viết, π=3.14).
(Kết quả làm tròn đến 2 chữ số sau dấu phẩy). 

```python
class Cylinder:
    def __init__(self,bankinh,chieucao):
        self.bk = bankinh
        self.cc = chieucao
    def getArea(self):
        return 2*3.14*self.bk**2 + 2*3.14*self.bk*self.cc
    def getVolume(self):
        return (3.14*self.bk**2) * self.cc
    def __str__(self):
        return f"({self.bk},{self.cc})"
bk,cc = map(int,input().split())

    
```

Bài 2
[Date]
Xây dựng lớp Date biểu diễn ngày tháng năm với hai phương
thức kiểm tra xem một ngày tháng nào đó có hợp lệ hay không
và phương thức tính ngày kế tiếp của ngày tháng năm nào đó
nếu nó là hợp lệ
Nhập vào một ngày tháng năm, (ngày tháng năm biểu diễn
dưới dạng dd/mm/yyyy). Nếu ngày tháng năm không tồn tại, in
ra dòng chữ INVALID. Nếu là hợp lệ in ra ngày kế tiếp của nó.
Đầu vào
Đầu vào từ bàn phim chứa dữ liệu ngày, tháng, năm dưới dạng
dd/mm/yyyy
.
Đầu ra
In ra màn hình ngày kế tiếp của ngày vừa nhập dưới dạng
dd/mm/yyyy
. Nếu ngày tháng không tồn tại, in ra INVALID

```python
class date:
    def __init__(self,d, m ,y):
        self.dd = d
        self.mm = m
        self.yy = y
    def checkVaild(self):
        maxday = -1
        match self.mm:
            case 1|3|5|7|8|10|12:
                maxday = 31
            case 4|6|9|11:
                maxday =30
            case 2:
                if (self.yy % 4 == 0 and self.yy % 100 == 0) or self.yy%400 ==0:
                    maxday = 29
                else:
                    maxday = 28
        return  1 <= self.dd <= maxday and self.yy > 0
    def getNextday(self):
        nextday = date(self.dd + 1,self.mm,self.yy)
        if nextday.checkVaild():
            return nextday
        nextday.dd = 1
        nextday.mm += 1
        if nextday.checkVaild():
            return nextday
        nextday.mm = 1
        nextday.yy += 1
        return nextday
    def __str__(self):
        return f"{self.dd:02d}/{self.mm:02d}/{self.yy}"
    
d,m,y = map(int, input().split("/"))
mydate1 = date(d,m,y)
print(mydate1)
print(mydate1.checkVaild())
if mydate1.checkVaild():
    nextday = mydate1.getNextday()
    print(nextday)
else:
    print("Ngay thang sai r")

```

Bài 3:[Point]
Xây dựng lớp điểm với hai thuộc tính x, y cùng các phương
thức kiểm tra xem có phải gốc tọa dộ không, có nằm trên trục
tung không, có nằm trên trục hoành không, tính khoảng cách từ
điểm đó tới một điểm khác
Đầu vào:
Nhập vào tọa độ một điểm M bất kỳ
Đầu ra:
Thông báo M có phải gốc tọa độ không hoặc có nằm trên trục
tung hoặc trục hoành không, khoảng cách từ M tới gốc tọa độ
là bao nhiêu (lấy 2 chữ số phần thập phân)

```python
import math
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def trucx(self):
        return self.y == 0
    def trucy(self):
        return self.x == 0
    def goc(self):
        return self.x == 0 and self.y == 0
    def kc(self):
        return math.sqrt((self.x)**2 + (self.y)**2)
a,b = map(int,input().split())
d1 = Point(a,b)
if d1.goc():
    print('nằm gốc')
elif d1.trucx():
    print('nằm trên trục x')
elif d1.trucy():
    print('nằm trục y')
print(f'kc:{d1.kc()}')

```


Bài 4:Xây dựng lớp Calculator (máy tính bỏ túi) với các phương thức
để thực hiện các phép toán 1. add (cộng), 2. subtract(trừ), 3.
product (nhân) và 4. divide(chia), 5. power(hàm mũ), 6. mod
(lấy phần dư) và phương thức 7. set_numbers(cập nhật 2 số
mới) nếu người dùng muốn tính toán trên 2 số khác.
Khởi tạo lớp là hai thuộc tính gồm 2 số cần tính, Mỗi phương
thức thực hiện một phép tính và trả về kết quả tương ứng
Tạo menu và cho người dùng tính toán tới khi gõ "yes", "YES"
hoặc "Yes" cho câu hỏi "Bạn có muốn thoát?" mới dừng.

```python
class Calculator:
    def __init__(self,a ,b):
        self.a = a
        self.b = b
    def add(self):
        return self.a +self.b
    def subtract(self):
        return self.a - self.b
    def product(self):
        return self.a * self.b
    def divide(self):
        if self.b == 0:
            return f'ko thể chia cho 0'
        return self.a/self.b
    def power(self):
        return self.a ** self.b
    def mod(self):
        return self.a  % self.b
    def set_numbers(self,a,b):
        self.a = a
        self.b = b
        return f'cap nhap 2 so moi'
print("--- CHÀO MỪNG ĐẾN VỚI MÁY TÍNH BỎ TÚI ---")
so_thu_nhat = float(input("Nhập số thứ nhất: "))
so_thu_hai = float(input("Nhập số thứ hai: "))

may_tinh = Calculator(so_thu_nhat, so_thu_hai)

while True:
    print("\n" + "="*30)
    print("MỜI BẠN CHỌN PHÉP TÍNH:")
    print("1. Cộng (Add)")
    print("2. Trừ (Subtract)")
    print("3. Nhân (Product)")
    print("4. Chia (Divide)")
    print("5. Hàm mũ (Power)")
    print("6. Chia lấy dư (Mod)")
    print("7. Nhập bộ số mới (Set numbers)")
    print("="*30)
    
    lua_chon = input("Nhập lựa chọn của bạn (1-7): ")
    match lua_chon:
        case "1":
            print(f"Kết quả: {may_tinh.a} + {may_tinh.b} = {may_tinh.add()}")
        case "2":
            print(f"Kết quả: {may_tinh.a} - {may_tinh.b} = {may_tinh.subtract()}")
        case "3":
            print(f"Kết quả: {may_tinh.a} * {may_tinh.b} = {may_tinh.product()}")
        case "4":
            print(f"Kết quả: {may_tinh.a} / {may_tinh.b} = {may_tinh.divide()}")
        case "5":
            print(f"Kết quả: {may_tinh.a} ^ {may_tinh.b} = {may_tinh.power()}")
        case "6":
            print(f"Kết quả: {may_tinh.a} % {may_tinh.b} = {may_tinh.mod()}")
        case "7":
            so_moi_1 = float(input("Nhập số thứ nhất mới: "))
            so_moi_2 = float(input("Nhập số thứ hai mới: "))
            may_tinh.set_numbers(so_moi_1, so_moi_2)
        case _:
            print("Lựa chọn không hợp lệ, vui lòng chọn từ 1 đến 7!")
    thoat = input("\nBạn có muốn thoát? ")
    if thoat in ["yes", "YES", "Yes"]:
        print("Đã tắt máy tính. Hẹn gặp lại!")
        break
        
```
