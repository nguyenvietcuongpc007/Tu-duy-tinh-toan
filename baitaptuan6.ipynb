Bài 4: Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key

```python
meb = {} 
i = input().split()
A = i.split()
for i in A:
    n, m = i.split(':')
    if n in meb:
        meb[n].append(m)
    else:
        meb[n] = [m]
print(meb)
```

Bài 5:Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ...,
'zeros': ...} (đếm số >0, <0, =0)

```python
A = list(map(int, input().split()))
meb = {}
d = 0
a = 0 
k = 0
for i in A:
    if i > 0:
        d += 1
    elif i < 0:
        a += 1
    else:
        k += 1
meb['positions'] = d
meb['negatives'] = a
meb['zezos'] = k
print(meb)
```

```python
A = list(map(int, input().split()))
res = {'positives': 0, 'negatives': 0, 'zeros': 0}

for i in A:
    if i > 0:
        res['positives'] += 1
    elif i < 0:
        res['negatives'] += 1
    else:
        res['zeros'] += 1

print(res)
```

Bài 9:Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
theo thứ tự từ điển tăng dần của key.

```python
def nhap_dict():
    d = {}
    s = input("Nhập chuỗi (ví dụ a:1 b:2): ").split()
    for item in s:
        key, val = item.split(':')
        d[key] = int(val)
    return d

a = nhap_dict()
b = nhap_dict()
res = a.copy()
for key, value in b.items():
    if key in res:
        res[key] += value
    else:
        sorted_keys = sorted(res.keys())
final_dict = {k: res[k] for k in sorted_keys}
print(final_dict)
```

Bài 10:Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.


```python
a = list(map(int, input().split()))
k = int(input())
res = []
for i in range(len(a)):
    for j in range(i+1,len(a)):
        if a[i] + a[j] == k and i < j:
            pair = (a[i], a[j])
            if pair not in res: 
                res.append(pair)
for i in range(len(res)):
    swap = False
    for j in range(len(res)-i-1):
        if res[j][0] > res[j+1][0]:
            res[j], res[j+1] = res[j+1], res[j]
            swap = True
    if not swap:
        break
print(res)

    

            
```

Bài 12:[Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list.
Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất
trong số đó. 


```python
a = list(map(int,input().split()))
meb = {}    
for i in a:
    if i in meb:
        meb[i] += 1
    else:
        meb[i] = 1
max_count = 0
res = 0
for k, v in meb.items():
    if v > max_count:
        max_count = v
        res = k
    elif v == max_count:
        if k < res:
            res = k
print(res)



    
```

Bài 13:[Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một
dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In
ra dictionary mới. 


```python
def nhap_dict():
    d = {}
    s = input("Nhập chuỗi (ví dụ a:1 b:2): ").split()
    for item in s:
        key, val = item.split(':')
        d[key] = int(val)
    return d

a = nhap_dict()
new_dict = {}
for k, v in a.items():
    new_dict[v] = k

print(new_dict)
```

Bài 14:Giao của hai danh sách]
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
phần tử chung đó, không trùng lặp.

```python
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))

# Chuyển sang set để lọc trùng và tìm giao
set1 = set(list1)
set2 = set(list2)

# Dùng toán tử & để lấy phần giao
res = list(set1 & set2)

print(res)
```
