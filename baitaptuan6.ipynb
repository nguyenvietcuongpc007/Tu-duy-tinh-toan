Bài 1:Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

```python
A = list(map(int,input().split()))
B = []
for i in A:
    if i not in B:
        B.append(i)
print(B)
    
```

```python
A = list(map(int, input().split()))

# dict.fromkeys(A) tạo ra các Key từ list A (loại trùng, giữ thứ tự)
# sau đó ta ép ngược lại thành list
B = list(dict.fromkeys(A))

print(B)
```

Bài 2:Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu
tới vị trí i), giữ nguyên thứ tự


```python
A = list(map(int, input().split()))
B = []
tong = 0
for i in A:
    tong = tong + i
    B.append(tong)
print(B)
```

Bài 3:Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp
xếp.)


```python
t = tuple(map(int,input().split()))
k = int(input())
n = len(t)
if k > n:
    k = k % n 
res = t[k:]+t[:k]
print(res)
```
