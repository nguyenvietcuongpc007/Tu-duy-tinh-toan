Bài 3:Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def snt(a):
    uoc = 0
    for i in range(1,int(a**0.5)+1):
        if a % i == 0:
            uoc = uoc + 1
    if uoc == 1: return True
    else: return False
```

Bài 4:VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def shh(a):
    sum = 0
    for i in range(1,int(a**0.5)+1):
        if a % i == 0:
            sum += i
            f  = a / i
            if f != i:
                sum += f
    if sum == a + a: return True
    else: return False
```

Bài 9:Viết hàm trả về tổng các chữ số của số nguyên dương n 

```python
def so(n):
    tong = 0
    while n !=0:
        a = n % 10
        tong += a
        n //= 10
    return tong
```

Bài 10:Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài. 


```python
def isIsomorphic(s, t):
    # Dùng 2 dictionary để kiểm tra 2 chiều
    map_st = {}
    map_ts = {}
    
    for i in range(len(s)):
        char_s = s[i]
        char_t = t[i]
        
        # Kiểm tra chiều từ S sang T
        if char_s in map_st:
            if map_st[char_s] != char_t:
                return False
        else:
            map_st[char_s] = char_t
            
        # Kiểm tra chiều ngược lại từ T sang S (để tránh 2 thằng trỏ vào 1)
        if char_t in map_ts:
            if map_ts[char_t] != char_s:
                return False
        else:
            map_ts[char_t] = char_s
            
    return True

# Test thử ví dụ của Cường
print(isIsomorphic("abca", "zbxz")) # True
print(isIsomorphic("abc", "aaa"))   # False
```
