Bài 9: Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    a = x1-x2
    b = y1 -y2
    c = x3 - x2
    d  = y3 -y2
    return a*d == b*c
```
