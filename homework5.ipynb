```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    a = x1-x2
    b = y1 -y2
    c = x3 - x2
    d  = y3 -y2
    return a*d == b*c
```

Bài 10:Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    # quad = [(x1,y1), (x2,y2), (x3,y3), (x4,y4)]
    # p = (px, py)
    
    def cross_product(a, b, p):
        # Tính (xB-xA)*(yP-yA) - (yB-yA)*(xP-xA)
        return (b[0] - a[0]) * (p[1] - a[1]) - (b[1] - a[1]) * (p[0] - a[0])

    signs = []
    for i in range(4):
        # Lấy đỉnh hiện tại và đỉnh kế tiếp (đỉnh 4 quay lại đỉnh 0)
        p1 = quad[i]
        p2 = quad[(i + 1) % 4]
        
        cp = cross_product(p1, p2, p)
        
        # Nếu cp == 0 thì điểm nằm trên biên cạnh đó
        if cp != 0:
            # Lưu lại dấu: 1 là dương (trái), -1 là âm (phải)
            signs.append(1 if cp > 0 else -1)
            
    # Nếu tất cả các dấu giống nhau (chỉ toàn 1 hoặc chỉ toàn -1)
    # Hoặc nếu điểm nằm trên biên (một số cp = 0 nên len(set) sẽ nhỏ hơn hoặc bằng 1)
    return len(set(signs)) <= 1

# Test thử ví dụ của Cường
quad = [(0,0), (4,0), (4,3), (0,3)]
p = (2,1)
print(point_in_quadrilateral(quad, p)) # Kết quả: True
```

Bài 11:Viết hàm trả về a XOR b.


```python
def so(a):
    if a == 0: return 0
    s = []
    while a > 0 :
        n = a % 2
        s.append(str(n))
        a //= 2
    return s[::-1]
def xor(a, b):
    str_a = so(a)
    str_b = so(b)
    
    # Bước quan trọng: Thêm số 0 vào đầu chuỗi ngắn hơn (Padding)
    max_len = max(len(str_a), len(str_b))
    str_a = str_a.zfill(max_len) # Dùng lệnh zfill mình vừa học nè
    str_b = str_b.zfill(max_len)
    
    res = ""
    for i in range(max_len):
        if str_a[i] == str_b[i]:
            res += "0"
        else:
            res += "1"
    return res
```

Bài 13:Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3)

```python
import math

def tinh_goc(A, B, C):
    # A, B, C là các tuple (x, y)
    # 1. Tính tọa độ vector BA và BC
    ba = (A[0] - B[0], A[1] - B[1])
    bc = (C[0] - B[0], C[1] - B[1])
    
    # 2. Tính tích vô hướng và độ dài vector
    dot_product = ba[0] * bc[0] + ba[1] * bc[1]
    len_ba = math.sqrt(ba[0]**2 + ba[1]**2)
    len_bc = math.sqrt(bc[0]**2 + bc[1]**2)
    
    # 3. Tính cos (tránh lỗi chia cho 0 nếu điểm trùng nhau)
    cos_abc = dot_product / (len_ba * len_bc)
    
    # Đảm bảo cos nằm trong khoảng [-1, 1] để không bị lỗi làm tròn
    cos_abc = max(-1, min(1, cos_abc))
    
    # 4. Tính góc và chuyển sang độ
    angle_rad = math.acos(cos_abc)
    angle_deg = math.degrees(angle_rad)
    
    return angle_deg

# Ví dụ: A(0, 1), B(0, 0), C(1, 0) -> Góc 90 độ
print(tinh_goc((0, 1), (0, 0), (1, 0)))
```

Bài 14:Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].


```python
def tim(a,b,c):
    max = a
    min = a
    if b >= max:
        max = b
    if b <= min:
        min = b
    if c >= max:
        max = c
    if c <= min:
        min = c
    return [max,min]
# Hoặc ngắn hơn nữa:
def tim_gon(a, b, c):
    return [max(a, b, c), min(a, b, c)]
```

Bài 15: Viết hàm tính hệ số nhị thức C(n, k).

```python
def to_hop_2(n, k):
    if k < 0 or k > n: return 0
    if k == 0 or k == n: return 1
    # Tính chất đối xứng: C(n, k) = C(n, n-k)
    # Chọn k nhỏ hơn để vòng lặp chạy ít hơn
    k = min(k, n - k)
    
    tu_so = 1
    mau_so = 1
    for i in range(k):
        tu_so *= (n - i)
        mau_so *= (i + 1)
    return tu_so // mau_so
```

Bài 18:Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá.


```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # 1. Cấu hình giá theo loại xe
    if car_type == 4:
        gia_mo_cua = 12000  # cho 1km đầu
        gia_tiep_theo = 15000 # từ km thứ 2 đến 30
        gia_duong_dai = 12000 # từ km 31 trở đi
    else: # Xe 7 chỗ
        gia_mo_cua = 15000
        gia_tiep_theo = 17000
        gia_duong_dai = 14000

    # 2. Tính tiền theo số km (Lũy tiến)
    if km <= 1:
        tien_xe = gia_mo_cua
    elif km <= 30:
        tien_xe = gia_mo_cua + (km - 1) * gia_tiep_theo
    else:
        tien_xe = gia_mo_cua + 29 * gia_tiep_theo + (km - 30) * gia_duong_dai

    # 3. Tính tiền chờ (Ví dụ: 3000đ/phút, miễn phí 5 phút đầu)
    tien_cho = 0
    if wait_minutes > 5:
        tien_cho = (wait_minutes - 5) * 3000

    # 4. Hệ số giờ cao điểm (Surge pricing)
    # Cao điểm: 7-9h và 16-19h
    he_so = 1.0
    if (7 <= start_hour <= 9) or (16 <= start_hour <= 19):
        he_so = 1.2 # Tăng 20% giá tiền xe
        
    tong_cong = (tien_xe * he_so) + tien_cho
    
    return round(tong_cong)

# Test thử: Xe 4 chỗ, đi lúc 17h (cao điểm), 10km, chờ 10 phút
print(f"Tiền cước: {smart_taxi_fare(4, 17, 10, 10)} VNĐ")
```

Bài 20:Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0

```python
#Tập hợp set là một loại "túi đựng" đặc biệt, nó sẽ tự động loại bỏ các phần tử trùng lặp.
def check_toan_hoc(x):
    da_xuat_hien = {}
    temp = x
    while temp > 0:
        chu_so = temp % 10
        if chu_so in da_xuat_hien:
            return False # Thấy lặp lại rồi!
        da_xuat_hien[chu_so] = True
        temp //= 10
    return True
```

```python
def check_khac_nhau(x):
    s = str(x)
    # So sánh độ dài chuỗi gốc và độ dài sau khi loại trùng bằng set
    return len(set(s)) == len(s)

def sum_1_to_n(n):
    # Điều kiện chặn đầu: Nếu n <= 0 trả về 0
    if n <= 0:
        return 0
    
    tong = 0
    # Chạy từ 1 đến n
    for i in range(1, n + 1):
        if check_khac_nhau(i):
            tong += i
            
    return tong
```
