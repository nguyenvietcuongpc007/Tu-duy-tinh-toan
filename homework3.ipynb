Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input())
print(" "*(h-1)+"*")
for i in range(h-1,1,-1):
    print(' '*(i-1)+'*',end='')
    print(' '*(2*(h-i)-1)+'*')
print('*'*(2*h-1))
```

```python
h = int(input())

for i in range(1, h + 1):
    # 1. Khoảng trắng đầu dòng luôn là h - i
    print(" " * (h - i), end="")
    
    if i == 1:
        # Dòng đỉnh
        print("*")
    elif i == h:
        # Dòng đáy
        print("*" * (2 * h - 1))
    else:
        # Dòng thân: In sao đầu + khoảng trắng giữa + sao cuối
        # Khoảng trắng giữa = 2*i - 3
        print("*" + " " * (2 * i - 3) + "*")
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
a = int(input())
s = 0
n = 0
while s <= a:
    n += 1
    s += 1/n
print(s,n)
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
max_val = None
min_val = None
while True:
    a = int(input())
    if a == - 1:
        break
    elif max_val is None or min_val is None :
        max_val = a
        min_val = a
    else :
        if a > max_val:
            max_val = a
        elif a < min_val:
            min_val =a
print(max_val,min_val)
```
