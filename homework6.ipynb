```python

import math
class Point:
    def __init__(self,toadox,toadoy):
        self.x = toadox
        self.y = toadoy
    def distance_to(self, pointY):
        dx = self.x - pointY.x
        dy = self.y - pointY.y
        res = math.sqrt(dx**2+dy**2)
        return res
class Tamgiac:
    def __init__(self,p1,p2,p3):
        self.x = p1
        self.y = p2
        self.z = p3
    def chuvi(self):
        xy = self.x.distance_to(self.y)
        yz = self.y.distance_to(self.z)
        zx = self.z.distance_to(self.x)
        return xy + yz +zx
p1 = Point(0,0)
p2 = Point(0,4)
p3 = Point(3,0)
tg = Tamgiac(p1,p2,p3)
print(tg.chuvi())

```

Bài 1:Nhâp vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
# Nhập số hàng n và số cột m
n, m = map(int, input("Nhập n và m: ").split())

# 1. Nhập ma trận
matrix = []
for i in range(n):
    # Nhập từng hàng và chuyển thành list số nguyên
    row = list(map(int, input(f"Nhập hàng {i+1}: ").split()))
    matrix.append(row)

# 2. In ma trận với định dạng 4 vị trí căn phải
print("\nKết quả ma trận:")
for row in matrix:
    for element in row:
        # Định dạng: %4d (kiểu cũ) hoặc {:4d} (kiểu mới)
        # Ở đây dùng f-string cho hiện đại: {element:4d}
        print(f"{element:4d}", end="") 
    print() # Xuống dòng sau khi in xong một hàng
```

Bài 2:Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo phụ của mảng

```python
n = int(input("Nhập kích thước ma trận vuông n: "))
matran = []
for i in range(n):
    row = list(map(int, input().split()))
    matran.append(row)
print("Đường chéo chính:", end=" ")
for i in range(n):
    print(matran[i][i], end=" ")
print()
print("Đường chéo phụ:", end=" ")
for i in range(n):
    print(matran[i][n-1-i], end=" ")
print()
```

Bài 4:Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list

```python
A = list(map(int,input().split()))
def max(a):
    max_value = a[0]
    dem = 0
    for i in range(len(a)):
        if a[i] > max_value:
            max_value = a[i]
            dem = i
    return dem
print(max(A))
```

Bài 6:Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
n = input().split()
meb = {}
for i in range(len(n)):
    if n[i] not in meb:
        meb[n[i]] = []
    else:
        meb[n[i]].append(i)
for x in meb.keys():
    meb[x] = tuple(meb[x])
print(meb)
```

Bài 9:Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def day_so_0_ve_dau(a):
    zeros = [x for x in a if x == 0]
    others = [x for x in a if x != 0]
    return zeros + others
a = list(map(int, input().split()))
print(day_so_0_ve_dau(a))
```

Bài 11:Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối

Bài 13:Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
lst = list(map(int,input().split()))
if len(set(lst)) != len(lst):
    print('vo danh')
elif lst == sorted(lst):
    print('tang')
elif lst == sorted(lst, reverse=True):
    print('giam')
else:
    print('vo danh')
```
