### Bài 3:

Nếu không có số dương nào thì trả về "invalid"

```python
lst = list(map(int,input().split()))
dem = 0
tong = 0
for i in lst:
    if i > 0:
        tong += i
        dem += 1
if dem == 0:
    print('invalid')
else:
    print(tong/dem)
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang

9 8 7 6 5 4 0: Giam

1 2 3 7 1 0 2: Vo danh

1 2 2 4 5 7: Vo danh

```python
lst = list(map(int,input().split()))
if len(set(lst)) != len(lst):
    print('vo danh')
elif lst == sorted(lst):
    print('tang')
elif lst == sorted(lst, reverse=True):
    print('giam')
else:
    print('vo danh')
```

```python
def is_ziczac(lst):
    n = len(lst)
    if n < 2: return True
    
    for i in range(n - 2):
        if (lst[i+1] - lst[i]) * (lst[i+2] - lst[i+1]) >= 0:
            return False
    return True
lst = list(map(int, input().split()))
if is_ziczac(lst):
    print("ziczac")
else:
    print("no ziczac")
    


```
